```python
# for making API requests
import requests
# showing Maps
import folium
import folium.plugins
# standard data processing and visualization
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rc('figure', figsize=(15, 4))
```

```python
df = pd.read_csv('got_rain.csv')
```

```python
df['time'] = pd.to_datetime(df['time'], format='mixed', utc=True)
df['date'] = df['time'].dt.date
df['time_of_day'] = df['time'].dt.time
df[['borough', 'intersection']] = df['name'].str.split(" - ", n=1, expand=True)
df[['street1', 'street2']] = df['intersection'].str.split("/", n=1, expand=True)
df['street1'] = df['street1'].str.strip()
df['street2'] = df['street2'].str.strip()
```

```python
df.to_csv('got_rain.csv', index=False)
```
