```python
import pandas as pd

# 1. Load the datasets
df_sensors = pd.read_csv('FloodNet__Sensor_Deployment_Metadata_20260313.csv')
df_events = pd.read_csv('FloodNet__Street_Flooding_Events_Measured_by_FloodNet_Sensors_20260313.csv')


# 2. Perform a left join to attach sensor metadata to the flood events
df_combined = pd.merge(
    df_events, 
    df_sensors, 
    on=['Sensor ID', 'Sensor Name'], 
    how='left'
)


df_combined.columns = df_combined.columns.str.lower()
rename_mapping = {
    'sensor name': 'name',
    'sensor id': 'sensor_id',
    'flood start datetime (gmt)': 'flood_start_gmt',
    'flood end datetime (gmt)': 'flood_end_gmt',
    'maximum flood depth (inches)': 'max_depth_inches',
    'time to maximum flood depth (minutes)': 'time_to_max_depth_min',
    'time to drain from peak (minutes)': 'time_to_drain_min',
    'totalduration_minutes': 'total_duration_min',
    'duration of flooding greater than 4 inches (minutes)': 'duration_over_4in_min',
    'duration of flooding greater than 12 inches (minutes)': 'duration_over_12in_min',
    'duration of flooding greater than 24 inches (minutes)': 'duration_over_24in_min',
    'time series depth values (inches)': 'depth_timeseries_inches',
    'time series depth timestamps (seconds)': 'timestamp_timeseries_sec',
    'date installed': 'date_installed',
    'tidally influenced': 'tidally_influenced',
    'date removed': 'date_removed',
    'street name': 'street_name',
    'borough': 'borough',
    'zipcode': 'zipcode',
    'communityboard': 'community_board',
    'council district': 'council_district',
    'census tract 2020': 'census_tract_2020',
    'nta': 'nta',
    'latitude': 'latitude',
    'longitude': 'longitude',
    'sensor ground height above local low point (inches)': 'sensor_ground_height_inches',
    'sensor location': 'sensor_location'
}

df_combined.rename(columns=rename_mapping, inplace=True)
df_combined['depth_inches'] = df_combined['max_depth_inches']
# 3. View the combined dataset and save it
print(df_combined.head())
df_combined.to_csv('combined_flood_data.csv', index=False)
```
