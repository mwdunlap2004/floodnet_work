```python
import optuna
from pathlib import Path

# Set the path to your database
current_location = Path.cwd().resolve()
if current_location.name in ["Finalized_Scripts", "Test_Scripts", "scripts"]:
    PROJECT_ROOT = current_location.parent
else:
    PROJECT_ROOT = current_location

DB = f"sqlite:///{PROJECT_ROOT}/Data_Files/floodnet_hpo.db"
study_names = ["log_ridge", "res_ann", "attn_lstm"]

print(f"Checking database at: {DB}\n")

for name in study_names:
    try:
        # Attempt to load the study
        study = optuna.load_study(study_name=name, storage=DB)
        
        # Count only completed trials (ignores crashed/pruned ones)
        completed_trials = len([t for t in study.trials if t.state == optuna.trial.TrialState.COMPLETE])
        
        print(f"✅ [{name}] loaded successfully!")
        print(f"   Completed Trials: {completed_trials}")
        print(f"   Best Target Val:  {study.best_value:.4f}")
        print(f"   Best Params:      {study.best_params}\n")
        
    except KeyError:
        print(f"❌ ERROR: Study '{name}' does not exist in the database.")
    except Exception as e:
        print(f"⚠️ ERROR loading '{name}': {e}")
```
